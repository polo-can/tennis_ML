# Tennis Match Prediction - Feature Engineering Pipeline

Pre-match prediction model using ATP data from 2015-2024.

**Key principles:**
- All features computed using ONLY data available before each match (no future leakage)
- 2015 used as warm-up year for building player histories; training data starts 2016+
- Temporal train/test split (not random) to simulate real prediction

**Features built:**
1. Elo ratings (overall + surface-specific)
2. Recent form (rolling win rates)
3. Head-to-head records
4. Surface performance
5. Fatigue / scheduling
6. Momentum / streaks
7. Tournament level performance
8. Historical serve statistics
9. Player profile (age, height, hand matchup, ranking trajectory)

## 1. Setup & Data Loading

```python
import pandas as pd
import numpy as np
from collections import defaultdict
from datetime import datetime, timedelta
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

DATA_DIR = Path('data')
np.random.seed(42)
```

```python
# Load and concatenate all match files (2015-2026)
dfs = []
for year in range(2015, 2027):
    fp = DATA_DIR / f'atp_matches_{year}.csv'
    if fp.exists():
        df_year = pd.read_csv(fp)
        dfs.append(df_year)
        print(f'{year}: {len(df_year)} matches')
    else:
        print(f'{year}: no file found, skipping')

df = pd.concat(dfs, ignore_index=True)
print(f'\nTotal: {len(df)} matches')
```

```python
# Parse dates and sort chronologically
df['tourney_date'] = pd.to_datetime(df['tourney_date'], format='%Y%m%d')
df = df.sort_values(['tourney_date', 'tourney_id', 'match_num']).reset_index(drop=True)

# Normalize surface: Carpet -> Hard
df['surface'] = df['surface'].replace('Carpet', 'Hard')

print(f'Date range: {df["tourney_date"].min()} to {df["tourney_date"].max()}')
print(f'Surfaces: {df["surface"].value_counts().to_dict()}')
print(f'Tournament levels: {df["tourney_level"].value_counts().to_dict()}')
```

```python
# Load player info for static features
players = pd.read_csv(DATA_DIR / 'atp_players.csv')
player_info = players.set_index('player_id')[['hand', 'height', 'dob', 'ioc']].to_dict('index')
print(f'Players loaded: {len(player_info)}')
```

## 2. Elo Rating System

```python
# K-factor by tournament level
K_FACTORS = {
    'G': 32,   # Grand Slam
    'M': 24,   # Masters 1000
    'A': 16,   # ATP 250/500
    'D': 12,   # Davis Cup
    'F': 24,   # Tour Finals
}
DEFAULT_K = 16
INITIAL_ELO = 1500.0


def elo_expected(player_elo, opponent_elo):
    """Expected win probability."""
    return 1.0 / (1.0 + 10.0 ** ((opponent_elo - player_elo) / 400.0))


def elo_update(player_elo, opponent_elo, won, k):
    """Return new Elo after a match result."""
    expected = elo_expected(player_elo, opponent_elo)
    return player_elo + k * (won - expected)


print('Elo system ready.')
print(f'K-factors: {K_FACTORS}')
```

## 3. Player State Storage

```python
# All player state is stored in dictionaries keyed by player_id.
# This is faster and simpler than using dataclasses for ~1000 players.

elo_overall = defaultdict(lambda: INITIAL_ELO)
elo_surface = defaultdict(lambda: defaultdict(lambda: INITIAL_ELO))  # [player_id][surface]

# Match history: list of dicts per player
match_history = defaultdict(list)

# Head-to-head: (player_id, opponent_id) -> {'wins': int, 'losses': int}
h2h_record = defaultdict(lambda: {'wins': 0, 'losses': 0})
# Surface-specific h2h: (player_id, opponent_id, surface) -> {'wins': int, 'losses': int}
h2h_surface_record = defaultdict(lambda: {'wins': 0, 'losses': 0})

# Surface records: player_id -> {surface -> {'wins': int, 'losses': int}}
surface_record = defaultdict(lambda: defaultdict(lambda: {'wins': 0, 'losses': 0}))

# Tournament level records
level_record = defaultdict(lambda: defaultdict(lambda: {'wins': 0, 'losses': 0}))

# Streak tracking
win_streak = defaultdict(int)
lose_streak = defaultdict(int)

# Total match counts
total_wins = defaultdict(int)
total_losses = defaultdict(int)

print('Player state storage initialized.')
```

## 4. Feature Extraction Functions

Each function extracts features for ONE player before a specific match, using only historical data.

```python
def get_elo_features(player_id, opponent_id, surface):
    """Elo ratings and expected win probability."""
    p_elo = elo_overall[player_id]
    o_elo = elo_overall[opponent_id]
    p_surf_elo = elo_surface[player_id][surface]
    o_surf_elo = elo_surface[opponent_id][surface]
    
    return {
        'elo': p_elo,
        'elo_opponent': o_elo,
        'elo_diff': p_elo - o_elo,
        'elo_expected': elo_expected(p_elo, o_elo),
        'surface_elo': p_surf_elo,
        'surface_elo_opponent': o_surf_elo,
        'surface_elo_diff': p_surf_elo - o_surf_elo,
        'surface_elo_expected': elo_expected(p_surf_elo, o_surf_elo),
    }
```

```python
def get_form_features(player_id, match_date):
    """Recent form: win rates over different windows."""
    history = match_history[player_id]
    if not history:
        return {
            'form_last10': np.nan,
            'form_last20': np.nan,
            'form_52w': np.nan,
            'form_weighted': np.nan,
            'matches_played': 0,
        }
    
    # Last N matches
    last10 = history[-10:]
    last20 = history[-20:]
    
    # Last 52 weeks
    cutoff_52w = match_date - timedelta(weeks=52)
    recent_52w = [m for m in history if m['date'] >= cutoff_52w]
    
    # Exponentially weighted form (decay = 0.9 per match, most recent first)
    decay = 0.9
    last30 = history[-30:]
    if last30:
        weights = np.array([decay ** i for i in range(len(last30) - 1, -1, -1)])
        results = np.array([m['won'] for m in last30], dtype=float)
        weighted_form = np.sum(weights * results) / np.sum(weights)
    else:
        weighted_form = np.nan
    
    def win_rate(matches):
        if not matches:
            return np.nan
        return np.mean([m['won'] for m in matches])
    
    return {
        'form_last10': win_rate(last10),
        'form_last20': win_rate(last20),
        'form_52w': win_rate(recent_52w),
        'form_weighted': weighted_form,
        'matches_played': len(history),
    }
```

```python
def get_h2h_features(player_id, opponent_id, surface):
    """Head-to-head record between the two players."""
    rec = h2h_record[(player_id, opponent_id)]
    total = rec['wins'] + rec['losses']
    
    surf_rec = h2h_surface_record[(player_id, opponent_id, surface)]
    surf_total = surf_rec['wins'] + surf_rec['losses']
    
    return {
        'h2h_wins': rec['wins'],
        'h2h_losses': rec['losses'],
        'h2h_total': total,
        'h2h_win_pct': rec['wins'] / total if total > 0 else np.nan,
        'h2h_surface_wins': surf_rec['wins'],
        'h2h_surface_losses': surf_rec['losses'],
        'h2h_surface_win_pct': surf_rec['wins'] / surf_total if surf_total > 0 else np.nan,
    }
```

```python
def get_surface_features(player_id, surface, match_date):
    """Surface-specific performance."""
    rec = surface_record[player_id][surface]
    total = rec['wins'] + rec['losses']
    
    # Recent surface form (last 52 weeks)
    cutoff = match_date - timedelta(weeks=52)
    recent = [m for m in match_history[player_id]
              if m['surface'] == surface and m['date'] >= cutoff]
    
    return {
        'surface_wins': rec['wins'],
        'surface_losses': rec['losses'],
        'surface_matches': total,
        'surface_win_pct': rec['wins'] / total if total > 0 else np.nan,
        'surface_recent_win_pct': np.mean([m['won'] for m in recent]) if recent else np.nan,
        'surface_recent_matches': len(recent),
    }
```

```python
def get_fatigue_features(player_id, match_date):
    """Fatigue and scheduling features."""
    history = match_history[player_id]
    if not history:
        return {
            'days_since_last': np.nan,
            'matches_last_7d': 0,
            'matches_last_14d': 0,
            'matches_last_30d': 0,
        }
    
    last_date = history[-1]['date']
    days_since = (match_date - last_date).days
    
    m7 = sum(1 for m in history if (match_date - m['date']).days <= 7)
    m14 = sum(1 for m in history if (match_date - m['date']).days <= 14)
    m30 = sum(1 for m in history if (match_date - m['date']).days <= 30)
    
    return {
        'days_since_last': days_since,
        'matches_last_7d': m7,
        'matches_last_14d': m14,
        'matches_last_30d': m30,
    }
```

```python
def get_momentum_features(player_id):
    """Current winning/losing streak."""
    return {
        'win_streak': win_streak[player_id],
        'lose_streak': lose_streak[player_id],
    }
```

```python
def get_level_features(player_id, tourney_level):
    """Win rate by tournament level."""
    rec = level_record[player_id][tourney_level]
    total = rec['wins'] + rec['losses']
    
    # Also compute Grand Slam specific win rate
    gs = level_record[player_id]['G']
    gs_total = gs['wins'] + gs['losses']
    
    return {
        'level_wins': rec['wins'],
        'level_losses': rec['losses'],
        'level_win_pct': rec['wins'] / total if total > 0 else np.nan,
        'gs_win_pct': gs['wins'] / gs_total if gs_total > 0 else np.nan,
    }
```

```python
def get_serve_features(player_id):
    """Rolling average serve statistics from past matches."""
    history = match_history[player_id]
    # Use last 30 matches with serve data
    with_serve = [m for m in history if m.get('svpt') and m['svpt'] > 0]
    recent = with_serve[-30:]
    
    if not recent:
        return {
            'avg_ace_rate': np.nan,
            'avg_df_rate': np.nan,
            'avg_first_serve_pct': np.nan,
            'avg_first_serve_win_pct': np.nan,
            'avg_second_serve_win_pct': np.nan,
            'avg_bp_saved_pct': np.nan,
        }
    
    # Weighted average (more weight to recent matches)
    n = len(recent)
    weights = np.array([0.95 ** (n - 1 - i) for i in range(n)])
    weights /= weights.sum()
    
    def wavg(key):
        vals = [m.get(key, np.nan) for m in recent]
        mask = ~np.isnan(vals)
        if not mask.any():
            return np.nan
        return np.average(np.array(vals)[mask], weights=weights[mask])
    
    return {
        'avg_ace_rate': wavg('ace_rate'),
        'avg_df_rate': wavg('df_rate'),
        'avg_first_serve_pct': wavg('first_serve_pct'),
        'avg_first_serve_win_pct': wavg('first_serve_win_pct'),
        'avg_second_serve_win_pct': wavg('second_serve_win_pct'),
        'avg_bp_saved_pct': wavg('bp_saved_pct'),
    }
```

```python
def get_player_profile(player_id, opponent_id, row, is_winner):
    """Static player features and matchup info."""
    if is_winner:
        rank = row.get('winner_rank', np.nan)
        rank_pts = row.get('winner_rank_points', np.nan)
        age = row.get('winner_age', np.nan)
        ht = row.get('winner_ht', np.nan)
        hand = row.get('winner_hand', 'U')
        opp_rank = row.get('loser_rank', np.nan)
        opp_rank_pts = row.get('loser_rank_points', np.nan)
        opp_age = row.get('loser_age', np.nan)
        opp_ht = row.get('loser_ht', np.nan)
        opp_hand = row.get('loser_hand', 'U')
    else:
        rank = row.get('loser_rank', np.nan)
        rank_pts = row.get('loser_rank_points', np.nan)
        age = row.get('loser_age', np.nan)
        ht = row.get('loser_ht', np.nan)
        hand = row.get('loser_hand', 'U')
        opp_rank = row.get('winner_rank', np.nan)
        opp_rank_pts = row.get('winner_rank_points', np.nan)
        opp_age = row.get('winner_age', np.nan)
        opp_ht = row.get('winner_ht', np.nan)
        opp_hand = row.get('winner_hand', 'U')
    
    # Ranking trajectory: compare current rank to rank 90 days ago
    history = match_history[player_id]
    rank_90d_ago = np.nan
    cutoff = row['tourney_date'] - timedelta(days=90)
    old_matches = [m for m in history if m['date'] <= cutoff and not np.isnan(m.get('rank', np.nan))]
    if old_matches:
        rank_90d_ago = old_matches[-1]['rank']
    
    # Hand encoding
    hand_map = {'R': 0, 'L': 1, 'U': 2}
    hand_enc = hand_map.get(hand, 2)
    opp_hand_enc = hand_map.get(opp_hand, 2)
    
    # Handedness matchup (R vs L is different from R vs R)
    same_hand = int(hand == opp_hand)
    is_lefty = int(hand == 'L')
    opp_is_lefty = int(opp_hand == 'L')
    
    return {
        'rank': rank,
        'rank_points': rank_pts,
        'age': age,
        'height': ht,
        'hand': hand_enc,
        'opp_rank': opp_rank,
        'opp_rank_points': opp_rank_pts,
        'opp_age': opp_age,
        'opp_height': opp_ht,
        'opp_hand': opp_hand_enc,
        'rank_diff': rank - opp_rank if not (np.isnan(rank) or np.isnan(opp_rank)) else np.nan,
        'rank_points_diff': rank_pts - opp_rank_pts if not (np.isnan(rank_pts) or np.isnan(opp_rank_pts)) else np.nan,
        'age_diff': age - opp_age if not (np.isnan(age) or np.isnan(opp_age)) else np.nan,
        'height_diff': ht - opp_ht if not (np.isnan(ht) or np.isnan(opp_ht)) else np.nan,
        'same_hand': same_hand,
        'is_lefty': is_lefty,
        'opp_is_lefty': opp_is_lefty,
        'rank_trajectory': rank - rank_90d_ago if not (np.isnan(rank) or np.isnan(rank_90d_ago)) else np.nan,
    }
```

## 5. Combined Feature Extraction & State Update

```python
def extract_all_features(player_id, opponent_id, surface, tourney_level, match_date, row, is_winner):
    """Extract ALL features for one player before a match."""
    features = {}
    features.update(get_elo_features(player_id, opponent_id, surface))
    features.update(get_form_features(player_id, match_date))
    features.update(get_h2h_features(player_id, opponent_id, surface))
    features.update(get_surface_features(player_id, surface, match_date))
    features.update(get_fatigue_features(player_id, match_date))
    features.update(get_momentum_features(player_id))
    features.update(get_level_features(player_id, tourney_level))
    features.update(get_serve_features(player_id))
    features.update(get_player_profile(player_id, opponent_id, row, is_winner))
    return features
```

```python
def compute_serve_stats(row, prefix):
    """Compute serve stat ratios for a player from a match row."""
    svpt = row.get(f'{prefix}_svpt', 0)
    if not svpt or np.isnan(svpt) or svpt == 0:
        return {}
    
    first_in = row.get(f'{prefix}_1stIn', 0) or 0
    second_attempts = svpt - first_in
    bp_faced = row.get(f'{prefix}_bpFaced', 0) or 0
    
    stats = {
        'svpt': svpt,
        'ace_rate': (row.get(f'{prefix}_ace', 0) or 0) / svpt,
        'df_rate': (row.get(f'{prefix}_df', 0) or 0) / svpt,
        'first_serve_pct': first_in / svpt if svpt > 0 else np.nan,
        'first_serve_win_pct': (row.get(f'{prefix}_1stWon', 0) or 0) / first_in if first_in > 0 else np.nan,
        'second_serve_win_pct': (row.get(f'{prefix}_2ndWon', 0) or 0) / second_attempts if second_attempts > 0 else np.nan,
        'bp_saved_pct': (row.get(f'{prefix}_bpSaved', 0) or 0) / bp_faced if bp_faced > 0 else np.nan,
    }
    return stats
```

```python
def update_state(player_id, opponent_id, won, surface, tourney_level, match_date, k_factor, row, prefix):
    """Update all player state after a match."""
    # 1. Update Elo
    opp_elo = elo_overall[opponent_id]
    elo_overall[player_id] = elo_update(elo_overall[player_id], opp_elo, won, k_factor)
    
    opp_surf_elo = elo_surface[opponent_id][surface]
    elo_surface[player_id][surface] = elo_update(
        elo_surface[player_id][surface], opp_surf_elo, won, k_factor
    )
    
    # 2. Compute serve stats for this match
    serve = compute_serve_stats(row, prefix)
    
    # 3. Get rank from this match
    if prefix == 'w':
        rank_val = row.get('winner_rank', np.nan)
    else:
        rank_val = row.get('loser_rank', np.nan)
    
    # 4. Append to match history
    entry = {
        'date': match_date,
        'won': won,
        'surface': surface,
        'tourney_level': tourney_level,
        'opponent_id': opponent_id,
        'rank': rank_val,
    }
    entry.update(serve)
    match_history[player_id].append(entry)
    
    # 5. Update h2h
    if won:
        h2h_record[(player_id, opponent_id)]['wins'] += 1
        h2h_surface_record[(player_id, opponent_id, surface)]['wins'] += 1
    else:
        h2h_record[(player_id, opponent_id)]['losses'] += 1
        h2h_surface_record[(player_id, opponent_id, surface)]['losses'] += 1
    
    # 6. Update surface record
    if won:
        surface_record[player_id][surface]['wins'] += 1
    else:
        surface_record[player_id][surface]['losses'] += 1
    
    # 7. Update level record
    if won:
        level_record[player_id][tourney_level]['wins'] += 1
    else:
        level_record[player_id][tourney_level]['losses'] += 1
    
    # 8. Update streaks
    if won:
        win_streak[player_id] += 1
        lose_streak[player_id] = 0
        total_wins[player_id] += 1
    else:
        lose_streak[player_id] += 1
        win_streak[player_id] = 0
        total_losses[player_id] += 1
```

## 6. Main Processing Loop

Iterate through all matches chronologically. For each match:
1. Extract features for both players (using only past data)
2. Store the feature row
3. Update all player state

```python
feature_rows = []

for idx, row in df.iterrows():
    winner_id = row['winner_id']
    loser_id = row['loser_id']
    surface = row['surface']
    tourney_level = row['tourney_level']
    match_date = row['tourney_date']
    k = K_FACTORS.get(tourney_level, DEFAULT_K)
    
    # --- EXTRACT FEATURES BEFORE UPDATING STATE ---
    winner_features = extract_all_features(
        winner_id, loser_id, surface, tourney_level, match_date, row, is_winner=True
    )
    loser_features = extract_all_features(
        loser_id, winner_id, surface, tourney_level, match_date, row, is_winner=False
    )
    
    # Store match metadata
    meta = {
        'match_date': match_date,
        'tourney_id': row['tourney_id'],
        'tourney_name': row['tourney_name'],
        'surface_cat': surface,
        'tourney_level_cat': tourney_level,
        'round': row['round'],
        'best_of': row['best_of'],
        'winner_id': winner_id,
        'loser_id': loser_id,
        'winner_name': row['winner_name'],
        'loser_name': row['loser_name'],
    }
    
    feature_rows.append({
        **meta,
        'w_features': winner_features,
        'l_features': loser_features,
    })
    
    # --- UPDATE STATE AFTER EXTRACTING FEATURES ---
    update_state(winner_id, loser_id, 1, surface, tourney_level, match_date, k, row, 'w')
    update_state(loser_id, winner_id, 0, surface, tourney_level, match_date, k, row, 'l')
    
    if (idx + 1) % 5000 == 0:
        print(f'Processed {idx + 1}/{len(df)} matches...')

print(f'Done! Processed {len(feature_rows)} matches.')
```

## 7. Dataset Assembly

For each match, randomly assign one player as `p1` and the other as `p2`.
Target: did `p1` win?

```python
rows = []

for match in feature_rows:
    # Randomly assign player1 and player2
    if np.random.random() < 0.5:
        p1_feat = match['w_features']
        p2_feat = match['l_features']
        target = 1  # p1 (winner) won
        p1_name = match['winner_name']
        p2_name = match['loser_name']
        p1_id = match['winner_id']
        p2_id = match['loser_id']
    else:
        p1_feat = match['l_features']
        p2_feat = match['w_features']
        target = 0  # p1 (loser) lost
        p1_name = match['loser_name']
        p2_name = match['winner_name']
        p1_id = match['loser_id']
        p2_id = match['winner_id']
    
    row_data = {
        'match_date': match['match_date'],
        'tourney_id': match['tourney_id'],
        'tourney_name': match['tourney_name'],
        'surface': match['surface_cat'],
        'tourney_level': match['tourney_level_cat'],
        'round': match['round'],
        'best_of': match['best_of'],
        'p1_id': p1_id,
        'p2_id': p2_id,
        'p1_name': p1_name,
        'p2_name': p2_name,
        'target': target,
    }
    
    # Add p1 features with 'p1_' prefix
    for k, v in p1_feat.items():
        row_data[f'p1_{k}'] = v
    
    # Add p2 features with 'p2_' prefix
    for k, v in p2_feat.items():
        row_data[f'p2_{k}'] = v
    
    # Add difference features (p1 - p2)
    for k in p1_feat:
        v1 = p1_feat[k]
        v2 = p2_feat[k]
        if isinstance(v1, (int, float)) and isinstance(v2, (int, float)):
            row_data[f'diff_{k}'] = v1 - v2
    
    rows.append(row_data)

dataset = pd.DataFrame(rows)
print(f'Dataset shape: {dataset.shape}')
print(f'Target distribution: {dataset["target"].value_counts().to_dict()}')
```

```python
# Filter out 2015 warm-up year
dataset = dataset[dataset['match_date'] >= '2016-01-01'].reset_index(drop=True)
print(f'Dataset after removing 2015 warm-up: {dataset.shape}')
print(f'Date range: {dataset["match_date"].min()} to {dataset["match_date"].max()}')
```

## 8. Temporal Train/Test Split

```python
# Train: 2016-2024, Test: 2025-2026
# (Previously was train 2016-2022, test 2023-2024)
# Adding 2023-2024 to training gives the model more recent data
train_mask = dataset['match_date'] < '2025-01-01'
test_mask = dataset['match_date'] >= '2025-01-01'

train_df = dataset[train_mask].copy()
test_df = dataset[test_mask].copy()

print(f'Train: {len(train_df)} matches ({train_df["match_date"].min()} to {train_df["match_date"].max()})')
print(f'Test:  {len(test_df)} matches ({test_df["match_date"].min()} to {test_df["match_date"].max()})')
print(f'\nTrain target: {train_df["target"].mean():.3f}')
print(f'Test target:  {test_df["target"].mean():.3f}')
```

## 9. Feature Analysis

```python
# Correlation of diff features with target
diff_cols = [c for c in feature_cols if c.startswith('diff_')]
corr = train_df[diff_cols + ['target']].corr()['target'].drop('target').sort_values(key=abs, ascending=False)

print('Top 20 diff features correlated with target:')
print(corr.head(20).to_string())
```

## 10. Save Processed Dataset

## 11. Model Training - Data Preparation

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score, log_loss, roc_auc_score, classification_report
from sklearn.model_selection import TimeSeriesSplit, RandomizedSearchCV
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

# Use only diff features + a few context features for the model
# Diff features capture the relative strength; individual p1/p2 features add redundancy
diff_cols = [c for c in feature_cols if c.startswith('diff_')]

# Add surface and best_of as context (one-hot encode surface)
train_model = train_df[diff_cols].copy()
test_model = test_df[diff_cols].copy()

# Add surface dummies
for surf in ['Hard', 'Clay', 'Grass']:
    train_model[f'surface_{surf}'] = (train_df['surface'] == surf).astype(int)
    test_model[f'surface_{surf}'] = (test_df['surface'] == surf).astype(int)

train_model['best_of'] = train_df['best_of']
test_model['best_of'] = test_df['best_of']

model_features = list(train_model.columns)
X_train = train_model.values
X_test = test_model.values
y_train = train_df['target'].values
y_test = test_df['target'].values

print(f'Model features: {len(model_features)}')
print(f'X_train: {X_train.shape}, X_test: {X_test.shape}')
print(f'Missing values in train: {np.isnan(X_train).sum()} ({np.isnan(X_train).mean()*100:.1f}%)')
```

```python
# Impute missing values for sklearn models (XGBoost handles NaN natively)
imputer = SimpleImputer(strategy='median')
X_train_imp = imputer.fit_transform(X_train)
X_test_imp = imputer.transform(X_test)

# Scale for logistic regression
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_imp)
X_test_scaled = scaler.transform(X_test_imp)

print('Imputation and scaling done.')
```

## 12. Logistic Regression (Baseline)

```python
lr = LogisticRegression(max_iter=1000, C=1.0, random_state=42)
lr.fit(X_train_scaled, y_train)

lr_probs = lr.predict_proba(X_test_scaled)[:, 1]
lr_preds = lr.predict(X_test_scaled)

print('=== Logistic Regression ===')
print(f'Accuracy:  {accuracy_score(y_test, lr_preds):.4f}')
print(f'Log Loss:  {log_loss(y_test, lr_probs):.4f}')
print(f'ROC AUC:   {roc_auc_score(y_test, lr_probs):.4f}')
print(f'\n{classification_report(y_test, lr_preds, target_names=["p2 wins", "p1 wins"])}')
```

## 13. Random Forest

```python
rf = RandomForestClassifier(
    n_estimators=500,
    max_depth=8,
    min_samples_leaf=20,
    max_features='sqrt',
    random_state=42,
    n_jobs=-1,
)
rf.fit(X_train_imp, y_train)

rf_probs = rf.predict_proba(X_test_imp)[:, 1]
rf_preds = rf.predict(X_test_imp)

print('=== Random Forest ===')
print(f'Accuracy:  {accuracy_score(y_test, rf_preds):.4f}')
print(f'Log Loss:  {log_loss(y_test, rf_probs):.4f}')
print(f'ROC AUC:   {roc_auc_score(y_test, rf_probs):.4f}')
print(f'\n{classification_report(y_test, rf_preds, target_names=["p2 wins", "p1 wins"])}')
```

## 14. Gradient Boosting with Hyperparameter Tuning

```python
# HistGradientBoostingClassifier handles NaN natively - use raw X_train/X_test
# Tune via TimeSeriesSplit cross-validation on training data

tscv = TimeSeriesSplit(n_splits=5)

param_grid = {
    'max_iter': [300, 500, 800],
    'max_depth': [4, 6, 8],
    'learning_rate': [0.01, 0.05, 0.1],
    'min_samples_leaf': [10, 20, 50],
    'max_leaf_nodes': [31, 63, 127],
    'l2_regularization': [0.0, 0.1, 1.0],
}

hgb_base = HistGradientBoostingClassifier(random_state=42)

search = RandomizedSearchCV(
    hgb_base,
    param_grid,
    n_iter=30,
    cv=tscv,
    scoring='neg_log_loss',
    random_state=42,
    n_jobs=-1,
    verbose=1,
)

print('Running HistGradientBoosting hyperparameter search (30 iterations, 5-fold TimeSeriesSplit)...')
search.fit(X_train, y_train)

print(f'\nBest params: {search.best_params_}')
print(f'Best CV log loss: {-search.best_score_:.4f}')
```

```python
# Evaluate best gradient boosting model on test set
hgb_best = search.best_estimator_

hgb_probs = hgb_best.predict_proba(X_test)[:, 1]
hgb_preds = hgb_best.predict(X_test)

print('=== Gradient Boosting (Tuned) ===')
print(f'Accuracy:  {accuracy_score(y_test, hgb_preds):.4f}')
print(f'Log Loss:  {log_loss(y_test, hgb_probs):.4f}')
print(f'ROC AUC:   {roc_auc_score(y_test, hgb_probs):.4f}')
print(f'\n{classification_report(y_test, hgb_preds, target_names=["p2 wins", "p1 wins"])}')
```

## 16. Feature Importance (Gradient Boosting)

```python
# Top 25 most important features (permutation importance would be more robust,
# but built-in importances are faster for a first look)
from sklearn.inspection import permutation_importance

# Use built-in feature importances (based on number of times feature is used in splits)
# HistGradientBoosting doesn't expose feature_importances_ by default in older sklearn,
# so we compute permutation importance on a subsample for speed
perm_result = permutation_importance(
    hgb_best, X_test, y_test, n_repeats=10, random_state=42, n_jobs=-1, scoring='accuracy'
)

importance = pd.Series(perm_result.importances_mean, index=model_features)
top25 = importance.sort_values(ascending=False).head(25)

fig, ax = plt.subplots(figsize=(10, 8))
top25.sort_values().plot(kind='barh', ax=ax)
ax.set_xlabel('Permutation Importance (accuracy decrease)')
ax.set_title('Top 25 Features - Gradient Boosting')
plt.tight_layout()
plt.savefig('feature_importance.png', dpi=100)
plt.show()

print('\nTop 25 features:')
for feat, imp in top25.items():
    print(f'  {feat:45s} {imp:.4f}')
```

## 17. Calibration Analysis

Are the predicted probabilities meaningful? A well-calibrated model should predict 70% win probability for matches where the player actually wins ~70% of the time.

```python
# Calibration plot for all models
from sklearn.calibration import calibration_curve

fig, ax = plt.subplots(figsize=(8, 8))

for name, probs_i in [('Logistic Regression', lr_probs), ('Random Forest', rf_probs), ('Gradient Boosting', hgb_probs)]:
    frac_pos, mean_pred = calibration_curve(y_test, probs_i, n_bins=10, strategy='uniform')
    ax.plot(mean_pred, frac_pos, 's-', label=name)

ax.plot([0, 1], [0, 1], 'k--', label='Perfect calibration')
ax.set_xlabel('Mean predicted probability')
ax.set_ylabel('Fraction of positives')
ax.set_title('Calibration Plot')
ax.legend()
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
plt.tight_layout()
plt.savefig('calibration.png', dpi=100)
plt.show()
```

## 18. Prediction Distribution & Confidence Analysis

```python
# How does accuracy vary with prediction confidence?
probs = hgb_probs
confidence = np.abs(probs - 0.5)  # 0 = uncertain, 0.5 = very confident

# Bucket by confidence and compute accuracy
bins = [0, 0.05, 0.10, 0.15, 0.20, 0.25, 0.30, 0.50]
labels = ['50-55%', '55-60%', '60-65%', '65-70%', '70-75%', '75-80%', '80%+']

conf_bin = pd.cut(confidence, bins=bins, labels=labels)
conf_df = pd.DataFrame({'confidence_bin': conf_bin, 'correct': (hgb_preds == y_test)})
conf_stats = conf_df.groupby('confidence_bin', observed=False).agg(
    accuracy=('correct', 'mean'),
    count=('correct', 'size'),
).reset_index()

print('Gradient Boosting accuracy by prediction confidence:')
print(conf_stats.to_string(index=False))
print(f'\nTotal test matches: {len(y_test)}')
```
